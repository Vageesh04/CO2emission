```python
train=data[:(int(len(data)*0.8))]
test=data[(int((len(data)*0.8))):]
print(train.shape)
print(test.shape)
```

```python
x_train=train.iloc[:, 3:12]
x_test=test.iloc[:, 3:12]
y_train=train.iloc[:, 12]
y_test=test.iloc[:, 12]
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
def model(x_train,y_train,LEARNING_RATE,ITERATION):
    m=y_train.size
    theta=np.zeros((9,1))
    y_train=np.array(y_train).reshape((m,-1))
    cost_list=[]
    
    for i in range(ITERATION):

        y_predicted=np.dot(x_train,theta)        
        cost=(1/(2*m)*np.sum(np.square(y_predicted-y_train)))        
        d_theta=(1/m)*np.dot(x_train.T,y_predicted-y_train)
        theta=theta-LEARNING_RATE*d_theta
        cost_list.append(cost)
          
    return theta, cost_list
```

```python
ITERATION=50000
LEARNING_RATE=0.0000001
theta, cost_list = model(x_train,y_train,LEARNING_RATE=LEARNING_RATE,ITERATION=ITERATION)
```

```python
rng=np.arange(0,ITERATION)
plt.plot(cost_list, rng)
plt.show()
```

```python
y_pred=np.dot(x_test,theta)
m=x_test.shape[0]


error=np.sqrt((1/m)*np.sum(np.square(y_pred-y_test)))
```
